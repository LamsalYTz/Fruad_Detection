```python
#1. Import Libraries
# ---------------------------
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#To ignore warnings
# ---------------------------
import warnings
warnings.filterwarnings('ignore')

#Load the dataset
# ---------------------------
df = pd.read_csv("../data/creditcard.csv")
# df.head()
# df.info()
# df.describe()
# df.shape

#Count Fraud vs Non-Fraud
# ---------------------------
# df['Class'].value_counts()

#Visualize Class Imbalance
#----------------------------
# plt.figure(figsize=(6,4))
# sns.countplot(df['Class'])
# plt.title("Fraud vs Non Fraud")
# plt.show()

#Percentage of Fraud Transactions
#--------------------------------
fraud_per = (df['Class'].value_counts()[1] / len(df)) * 100
print("Fruad Percentage", fraud_per)

#Visualize Transaction Amount Distribution
#-------------------------------
plt.figure(figsize=(20, 8))
sns.histplot(df['Amount'], kde=True, bins=50)
plt.title("Transaction Amount Distribution")
plt.show()

#Correlation Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(), cmap = 'plasma', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
